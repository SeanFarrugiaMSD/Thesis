```python
import numpy as np
import copy

coord_loc_list = [
    [
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
        [(632, 164),	(608, 211),	(585, 211),	(538, 234),	(585, 281),	(655, 211),	(655, 258),	(632, 281),	(585, 305),	(608, 399),	(608, 469),	(632, 305),	(679, 375),	(679, 469),	(632, 164),	(632, 164),	(608, 164),	(655, 164),	(23, 0)],
    ],
    [
        [(444, 187),	(444, 211),	(421, 234),	(398, 258),	(421, 305),	(468, 211),	(491, 281),	(491, 305),	(444, 328),	(444, 422),	(468, 493),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
        [(444, 187),	(444, 211),	(421, 211),	(398, 258),	(421, 305),	(468, 211),	(491, 281),	(491, 305),	(444, 328),	(444, 422),	(468, 493),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
        [(444, 187),	(444, 211),	(421, 211),	(398, 258),	(421, 305),	(491, 211),	(491, 258),	(491, 305),	(444, 328),	(444, 422),	(468, 493),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
        [(444, 187),	(444, 211),	(421, 211),	(398, 258),	(421, 305),	(491, 211),	(491, 258),	(491, 305),	(444, 328),	(444, 422),	(468, 516),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
        [(444, 187),	(444, 211),	(421, 211),	(398, 258),	None,	(468, 211),	(491, 258),	(491, 305),	(444, 328),	(444, 422),	(468, 493),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
        [(444, 187),	(444, 211),	(421, 211),	(398, 258),	(421, 305),	(491, 211),	(491, 258),	(491, 305),	(444, 328),	(444, 422),	(468, 516),	(491, 328),	(491, 422),	(515, 493),	(444, 164),	(444, 164),	(421, 187),	(468, 187),	(23, 0)],
    ],
    [
        [None,	(983, 845),	(983, 868),	(1030, 939),	(936, 962),	(983, 845),	None,	None,	(1077, 915),	(1006, 1009),	None,	(1123, 892),	(1077, 1033),	None,	None,	None,	None,	None,	(23, 0)],
        [None,	(983, 845),	(983, 845),	(1030, 939),	(936, 962),	(983, 845),	(1030, 939),	None,	(1077, 915),	(1006, 1009),	None,	(1123, 915),	(1077, 1033),	None,	None,	None,	None,	None, (23, 0)],
        [None,	(983, 845),	(983, 845),	(1030, 939),	(936, 962),	(960, 845),	(889, 892),	(866, 939),	(1100, 915),	(1006, 1009),	None,	(1123, 915),	(1077, 1033),	None,	None,	None,	None,	None,(23, 0)],
        [None,	(983, 845),	(983, 845),	(1030, 939),	(936, 962),	(983, 845),	(1030, 939),	None,	(1100, 915),	(1006, 1009),	None,	(1123, 892),	(1077, 1033),	None,	None,	None,	None,	(913, 892),	(23, 0)],
        [None,	(983, 845),	(983, 845),	(1030, 939),	(936, 962),	(983, 845),	(1030, 939),	None,	(1100, 915),	(1006, 1009),	None,	(1123, 892),	(1077, 1033),	None,	None,	None,	None,	None,	(23, 0)],
        [None,	(983, 845),	(1006, 821),	(1030, 939),	(960, 939),	(983, 868),	(1030, 939),	(960, 962),	(1100, 915),	(1006, 1009),	None,	(1123, 892),	(1077, 1033),	None,	None,	None,	None,	(913, 868),	(23, 0)],
    ],
    [
        [None,	(1053, 610),	(983, 586),	(960, 680),	(936, 774),	(1100, 633),	(1100, 751),	None,	(1030, 563),	None,	None,	(1100, 563),	(1123, 751),	None,	None,	None,	None,	None,	(23, 0)],
        [None,	(1053, 586),	(983, 586),	(936, 680),	(936, 751),	(1100, 586),	(1100, 727),	None,	(1030, 751),	None,	None,	(1100, 751),	(1123, 751),	None,	None,	None,	(1006, 586),	(1100, 586),	(23, 0)],
        [None,	(1030, 586),	(960, 586),	(936, 657),	(936, 751),	(1100, 586),	(1100, 727),	None,	(1006, 563),	(1030, 727),	(1006, 774),	(1100, 540),	(1100, 751),	None,	None,	None,	(1006, 586),	None,	(23, 0)],
        [None,	(1030, 586),	(960, 586),	(936, 657),	(936, 751),	(1100, 563),	None,	None,	(1006, 540),	(1100, 727),	None,	(1077, 540),	(1100, 727),	(1006, 774),	None,	None,	None,	None,	(23, 0)],
        [None,	(1030, 633),	(960, 586),	(913, 680),	(936, 774),	(1100, 657),	(1077, 751),	(1006, 774),	(1006, 540),	(936, 680),	(913, 751),	(1077, 540),	(1100, 727),	(1100, 845),	None,	None,	(983, 657),	(1030, 657),	(23, 0)],
        [None,	(1030, 610),	(960, 586),	(913, 680),	(936, 751),	(1100, 633),	(1077, 727),	(1006, 774),	(983, 540),	(913, 680),	(913, 727),	(1053, 563),	(1100, 727),	None,	None,	None,	(1006, 633),	(1030, 657),	(23, 0)],
    ],
    [
        [(866, 0),	(866, 46),	(796, 70),	(796, 164),	(796, 234),	(913, 46),	(960, 140),	(960, 211),	(842, 211),	(889, 352),	(913, 469),	(913, 211),	(913, 352),	(913, 469),	None,	None,	(842, 0),	None,	(23, 0)],
        [(866, 0),	(866, 46),	(819, 70),	(796, 164),	(796, 234),	(936, 46),	(960, 140),	(960, 234),	(842, 234),	(889, 352),	(936, 469),	(913, 211),	(913, 352),	(913, 469),	None,	None,	None,	None,	(23, 0)],
        [None,	(866, 46),	(819, 70),	(796, 164),	(796, 234),	(936, 46),	(960, 140),	(960, 234),	(842, 234),	(889, 352),	(913, 469),	(913, 211),	(913, 352),	(913, 469),	None,	None,	None,	None,	(23, 0)],
        [(866, 0),	(866, 46),	(819, 70),	(796, 164),	(796, 258),	(936, 46),	(960, 140),	(960, 211),	(842, 234),	(866, 352),	(936, 469),	(913, 234),	(913, 352),	(913, 493),	(866, 0),	(889, 0),	(842, 0),	(889, 0),	(23, 0)],
        [None,	(866, 46),	(819, 70),	(796, 164),	(796, 258),	(936, 46),	(960, 140),	(960, 234),	(842, 234),	(866, 352),	(913, 469),	(913, 234),	(913, 352),	(913, 469),	None,	None,	None,	None,	(23, 0)],
        [(866, 23),	(866, 70),	(819, 70),	(796, 164),	(796, 258),	(936, 46),	(960, 140),	(960, 234),	(842, 234),	(866, 375),	(936, 469),	(913, 234),	(913, 352),	(913, 493),	(866, 23),	(889, 23),	(842, 0),	(889, 0),	(23, 0)],
        [None,	(866, 70),	(819, 70),	(796, 164),	(796, 258),	(936, 46),	(960, 164),	(960, 234),	(842, 234),	(866, 375),	(866, 399),	(913, 234),	(913, 375),	(913, 493),	None,	None,	None,	None,	(23, 0)],
        [(866, 46),	(866, 70),	(819, 70),	(796, 164),	(796, 258),	(936, 70),	(960, 164),	(960, 234),	(842, 234),	(866, 375),	(913, 469),	(936, 234),	(913, 375),	(913, 469),	(866, 23),	(889, 23),	(842, 23),	(913, 23),	(23, 0)],
        [None,	(889, 70),	(819, 70),	(796, 187),	(796, 258),	(936, 70),	(960, 164),	(983, 258),	(842, 258),	(866, 399),	(889, 469),	(936, 258),	(913, 375),	(913, 469),	None,	None,	None,	None,	(23, 0)],
    ],
    [
        [None,	None,	None,	None,	None,	None,	None,	None,	(1123, 0),	(1123, 70),	(1100, 140),	(1170, 0),	(1170, 93),	(1170, 164),	None,	None,	None,	None,	(1053, 352)],
        [None,	None,	None,	None,	None,	None,	None,	(1194, 23),	(1123, 0),	(1123, 70),	(1123, 140),	(1170, 0),	(1170, 70),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	(1217, 0),	(1123, 0),	(1123, 70),	(1123, 140),	(1170, 23),	(1194, 93),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	None,	(1147, 46),	(1123, 93),	(1123, 164),	(1170, 23),	(1170, 93),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	(1194, 0),	(1123, 0),	(1123, 70),	(1123, 164),	(1194, 0),	(1170, 93),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	(1147, 23),	None,	None,	(1194, 23),	(1170, 23),	(1147, 93),	(1147, 164),	(1194, 23),	(1170, 93),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	(1194, 0),	(1147, 23),	(1123, 93),	(1123, 164),	(1194, 23),	(1170, 93),	(1170, 164),	None,	None,	None,	None,	(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	(1194, 0),	(1147, 0),	(1123, 117),	(1123, 211),	(1194, 23),	(1194, 93),	(1170, 164),	None,	None,	None,	None,(23, 0)],
        [None,	None,	None,	None,	None,	None,	None,	None,	(1194, 0),	(1194, 93),	(1170, 164),	(1147, 0),	(1123, 117),	(1123, 211),	None,	None,	None,	None,	(23, 0)]
    ]
]

def median_aug(list_keys):

    count = 0
    list_compare = []

    while list_compare != list_keys:

        count += 1
        list_compare = copy.deepcopy(list_keys)

        for person in range(len(list_keys)):
            for frame_keypoints in range(len(list_keys[person])):
                for spec_keypoint in range(len(list_keys[person][frame_keypoints])):
                    if list_keys[person][frame_keypoints][spec_keypoint] == None:

                        #Find previous + next and make average
                        for key_count in range(1,10):

                            if frame_keypoints-key_count >= 0:
                                prev_key = list_keys[person][frame_keypoints-key_count][spec_keypoint]
                            else:
                                break

                            if frame_keypoints+key_count <= len(list_keys[person])-1:
                                next_key = list_keys[person][frame_keypoints+key_count][spec_keypoint]
                            else:
                                break

                            if prev_key != None and next_key != None:

                                projected_x = int((prev_key[0] + next_key[0]) / 2)
                                projected_y = int((prev_key[1] + next_key[1]) / 2)

                                projected_key = (projected_x,projected_y)
                                list_keys[person][frame_keypoints][spec_keypoint] = projected_key
                                break

    if count == 1:
        return True
    else:
        return False

def linear_aug(list_keys):

    list_compare = copy.deepcopy(list_keys)

    for person in range(len(list_keys)):
        for frame_keypoints in range(len(list_keys[person])):
            for spec_keypoint in range(len(list_keys[person][frame_keypoints])):
                if list_keys[person][frame_keypoints][spec_keypoint] == None:

                    if frame_keypoints-2 >= 0:
                        prev_key_2 = list_keys[person][frame_keypoints-2][spec_keypoint]
                    else:
                        prev_key_2 = None

                    if frame_keypoints-1 >= 0:
                        prev_key = list_keys[person][frame_keypoints-1][spec_keypoint]
                    else:
                        prev_key = None

                    try:
                        next_key = list_keys[person][frame_keypoints+1][spec_keypoint]
                    except:
                        next_key = None

                    try:
                        next_key_2 = list_keys[person][frame_keypoints+2][spec_keypoint]
                    except:
                        next_key_2 = None


                    #If still None
                    if coord_loc_list[person][frame_keypoints][spec_keypoint] == None:

                        #Find previous x2
                        if prev_key != None and prev_key_2 != None:

                            projected_x = int(prev_key[0] + (prev_key[0] - prev_key_2[0]))
                            projected_y = int(prev_key[1] + (prev_key[1] - prev_key[1]))

                            projected_key = (projected_x,projected_y)
                            coord_loc_list[person][frame_keypoints][spec_keypoint] = projected_key


                        #Find next x2
                        if next_key != None and next_key_2 != None:

                            projected_x = int(next_key[0] - (next_key_2[0] - next_key[0]))
                            projected_y = int(next_key[1] - (next_key_2[1] - next_key[1]))

                            projected_key = (projected_x,projected_y)
                            coord_loc_list[person][frame_keypoints][spec_keypoint] = projected_key

    if list_compare == list_keys:
        return True
    else:
        return False

def sort_data(list_keys):

    csv_list = []

    for x in range(len(list_keys)):
        for y in range(0, len(list_keys[x]),3):
            try:
                list_output_keypoints = []
                for z in range(len(list_keys[x][y])):
                    for frame_count in range(3):
                        for coord_count in range(2):
                            list_output_keypoints.append(list_keys[x][y + frame_count][z][coord_count])

                csv_list.append(["violent"] + list_output_keypoints)
            except:
                runNothing()

    return csv_list

def runNothing():
    return None

#def dummy_replace(list_keys):

def cull_keys(list_keys):
    removed_dub_list = []
    [removed_dub_list.append(x) for x in list_keys if x not in removed_dub_list]

    return removed_dub_list

init_count = 0
final_count = 0

for person in range(len(coord_loc_list)):
    for frame_keypoints in range(len(coord_loc_list[person])):
        for spec_keypoint in range(len(coord_loc_list[person][frame_keypoints])):
            if coord_loc_list[person][frame_keypoints][spec_keypoint] == None:
                init_count += 1

while(True):

    med_res = median_aug(coord_loc_list)
    line_res = linear_aug(coord_loc_list)

    if med_res and line_res:
        break

for person in range(len(coord_loc_list)):
    for frame_keypoints in range(len(coord_loc_list[person])):
        for spec_keypoint in range(len(coord_loc_list[person][frame_keypoints])):
            if coord_loc_list[person][frame_keypoints][spec_keypoint] == None:
                final_count += 1

print("None Initially: " + str(init_count))
print("None Remaining: " + str(final_count))

for x in coord_loc_list:
    for y in x:
        print(y)

print("-----------------")

sorted_coord_list = sort_data(coord_loc_list)

for x in sorted_coord_list:
    print(x)

print("-----------------")

unique_coord_list = cull_keys(sorted_coord_list)

for x in unique_coord_list:
    print(x)
```
